```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randn(7, 3))

df.iloc[:4, 1] = np.nan # this iloc function mean 4th row column 1st convert NaN
df.iloc[:2, 2] = np.nan # this iloc function mean 2nd row column 2nd convert NaN
display(df)
print(df)

df.replace(np.nan, -999)
display(df)
print(df)

df = df.replace(-999, -9)
display(df)

df = df.replace([-999, 9], [np.nan, 0])
display(df)
print(df)
```
