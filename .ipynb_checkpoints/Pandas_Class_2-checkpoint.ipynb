# Apply(function)

```python
def fee_update(x):
    if x == "A.I":
        return 1100
    elif x == "AIOT":
        return 1500
    elif x == "CNC":
        return 2000
    else:
        return 0
    
#dataframe.columnName.apply(user_define_function)
df1.Course.apply(fee_update)
```

```python
df1['updated fee'] = df1.Course.apply(fee_update)
df1
```

```python
def custum_fn(x1, *l1):
    if x1 == "A.I":
        return np.sum(l1)
    elif x1 == "AIOT":
        return np.product(l1)
    elif x1 == "CNC":
        return np.cumsum(l1)
    else:
        return 0

```
