```python
import pandas as pd
import numpy as np
from numpy import nan as NA
```

```python
data = pd.DataFrame({'k1' : ['one', 'two'] * 3 + ['two'],
                    'k2' : [1, 1, 2, 3, 3, 4, 4]})

display(data)
data.duplicated()

# drop duplicates mean drop complete row
data.drop_duplicates()
display(data)

data['v1'] = ['one', 'two', 'one', 'four', 'one', 'six', 'two']
display(data)

data = data.drop_duplicates(['k1', 'v1'])
data
```
