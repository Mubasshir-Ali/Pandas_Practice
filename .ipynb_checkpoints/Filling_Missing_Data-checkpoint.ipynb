# filling of data

```python
import pandas as pd
import numpy as np
from numpy import nan as NA
```

```python
df = pd.DataFrame(np.random.randn(7, 3))
#print(df)

# iloc function using column or rows location data will be update
df.iloc[:4, 1] = NA # this iloc function mean 4th row column 1st convert NA
df.iloc[:2, 2] = NA # this iloc function mean 2nd row column 2nd convert NA
display(df)
print(df)

#calling fillna with a dict, you can use a different fill value for
df.fillna(1.5)
display(df)

# column number 1 will be filled with 0.5
# column number 2 will be filled with 0
df.fillna({1: 0.5, 2: 0})

# fill not values from left column (axis = 0 means from top row)
df.fillna(method = 'ffill', axis = 1)
df.fillna(method = 'ffill', limit = 1, axis = 1)
```

```python
df = pd.DataFrame(np.random.randn(7, 3))
#print(df)

# iloc function using column or rows location data will be update
df.iloc[:4, 1] = NA # this iloc function mean 4th row column 1st convert NA
display(df)
print(df)
```

```python
df = pd.DataFrame(np.random.randn(7, 3))
#print(df)

# iloc function using column or rows location data will be update
df.iloc[:2, 2] = NA # this iloc function mean 2nd row column 2nd convert NA
display(df)
print(df)
```

```python
df = pd.DataFrame(np.random.randn(7, 3))
#print(df)

#calling fillna with a dict, you can use a different fill value for
df.fillna(1.5)
display(df)
```

```python
df = pd.DataFrame(np.random.randn(7, 3))

# column number 1 will be filled with 0.5
# column number 2 will be filled with 0
df.fillna({1: 0.5, 2: 0})
```

```python
df = pd.DataFrame(np.random.randn(7, 3))

# fill not values from left column (axis = 0 means from top row)
df.fillna(method = 'ffill', axis = 1)
```

```python
df = pd.DataFrame(np.random.randn(7, 3))

# fill not values from left column (axis = 0 means from top row)
df.fillna(method = 'ffill', limit = 1, axis = 1)
```

```python
df = pd.DataFrame(np.random.randn(7, 3))
print(df)
```

```python
df = pd.DataFrame(np.random.randn(7, 3))

# fill not values from left rows (axis = 0 means from top row)
df.fillna(method = 'ffill', axis = 0)
df.fillna(method = 'ffill', limit = 1, axis = 0)
```

```python
df = pd.DataFrame(np.random.randn(7, 3))

# fill not values from left rows (axis = 0 means from top row)
df.fillna(method = 'ffill', axis = 0)
df.fillna(method = 'ffill', limit = 5, axis = 0)
```

```python
df = pd.DataFrame(np.random.randn(7, 3))

# fill not values from left columns (axis = 0 means from top row)
df.fillna(method = 'ffill', axis = 1)
df.fillna(method = 'ffill', limit = 5, axis = 1)
```
